```python
# import necessary library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
# 2. Sample data
x1 = 3
x2 = 2
y_actual = 17  # Actual value based on y = 3 * x1 + 4 * x2
```

```python
# 3. Initialize random weights between 1 and 10
np.random.seed(42)  # For reproducibility
w1 = np.random.uniform(1, 10)  # weight for x1
w2 = np.random.uniform(1, 10)  # weight for x2
```

```python
# 4. Hyperparameters
learning_rate = 0.01  # Learning rate for weight update
epochs = 20  # Number of iterations
# Lists to store error and predicted values for visualization
errors = []
ypreds = []
```

```python
# 5. Training loop (Feed-Forward and Backpropagation)
for epoch in range(epochs):
    # Feed-forward: compute the predicted output
    y_pred = x1 * w1 + x2 * w2  # Linear model: y_pred = x1 * w1 + x2 * w2

    # Compute squared error
    error = (y_actual - y_pred) ** 2
    errors.append(error)
    ypreds.append(y_pred)

    # Backpropagation: Compute gradients and update weights
    d_error_w1 = -2 * x1 * (y_actual - y_pred)  # Gradient of error with respect to w1
    d_error_w2 = -2 * x2 * (y_actual - y_pred)  # Gradient of error with respect to w2

    # Update weights using gradient descent
    w1 -= learning_rate * d_error_w1
    w2 -= learning_rate * d_error_w2
```

```python
# 6. Plotting the error progression and predicted output progression
plt.figure(figsize=(12, 6))

# Plot Error progression
plt.subplot(1, 2, 1)
plt.plot(range(epochs), errors, label="Error (Squared)")
plt.title('Error Progression Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.grid(True)

# Plot Predicted values progression
plt.subplot(1, 2, 2)
plt.plot(range(epochs), ypreds, label="Predicted y", color='orange')
plt.title('Predicted Output Progression Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Predicted Output (y_pred)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Final output after training
print(f"Final predicted value: {ypreds[-1]:.2f}")
print(f"Final weights: w1={w1:.2f}, w2={w2:.2f}")
```
